```python
from astropy.table import Table

import seaborn
```

```python
photometry = Table.read('data/HD38451.ecsv')
```

```python
mag = seaborn.scatterplot(
    x=photometry['timestamp'],
    y=photometry['MAG_AUTO'],
    hue=photometry['band'],
)
mag.invert_yaxis()
```

```python
flux = seaborn.scatterplot(
    x=photometry['timestamp'],
    y=photometry['FLUX_AUTO'],
    hue=photometry['band'],
)
```
